```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados das corridas com os pilotos e tempos de volta
dados_corridas = {
    'Piloto': ['Sam Bird', 'Nick Cassidy', 'Lucas Di Grassi', 'Jean-Éric Vergne', 'Stoffel Vandoorne'],
    'Equipe': ['Jaguar Racing', 'Envision Virgin Racing', 'ROKiT Venturi Racing', 'DS Techeetah', 'Mercedes-EQ Formula E Team'],
    'Tempo Volta 1': [75, 78, 76, 80, 77],
    'Tempo Volta 2': [74, 79, 75, 81, 76],
    'Tempo Volta 3': [73, 77, 74, 82, 75]
}

df_corridas = pd.DataFrame(dados_corridas)
df_corridas['Tempo Médio'] = df_corridas[['Tempo Volta 1', 'Tempo Volta 2', 'Tempo Volta 3']].mean(axis=1)
print("\nEstatísticas de Corridas:")
print(df_corridas)

# Função para armazenar estatísticas em CSV
def armazenar_estatisticas(df):
    df.to_csv('estatisticas_corridas.csv', index=False)

armazenar_estatisticas(df_corridas)

def ler_estatisticas():
    df_lido = pd.read_csv('estatisticas_corridas.csv')
    print("\nEstatísticas Lidas do Arquivo CSV:")
    print(df_lido)
    
    plt.figure(figsize=(10, 6))
    plt.bar(df_lido['Piloto'], df_lido['Tempo Médio'], color='blue')
    plt.xlabel('Piloto')
    plt.ylabel('Tempo Médio (s)')
    plt.title('Tempo Médio por Piloto')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

ler_estatisticas()

# Função para calcular classificação com base nas posições dos pilotos
def calcular_classificacao(pilotos_posicoes):
    pontos_por_posicao = {
        1: 25,
        2: 18,
        3: 15,
        4: 12,
        5: 10,
        6: 8,
        7: 6,
        8: 4,
        9: 2,
        10: 1
    }
    
    classificacao = {}
    for piloto, posicao in pilotos_posicoes.items():
        if posicao in pontos_por_posicao:
            classificacao[piloto] = pontos_por_posicao[posicao]
        else:
            classificacao[piloto] = 0  # Pilotos fora do top 10 não ganham pontos
    return classificacao

def exibir_ranking(classificacao):
    ranking = sorted(classificacao.items(), key=lambda x: x[1], reverse=True)
    print("\nClassificação Final:")
    for i, (piloto, pontos) in enumerate(ranking, start=1):
        print(f"{i}º Lugar: {piloto} - {pontos} pontos")

# Posições dos pilotos nas corridas
pilotos_posicoes = {
    'Sam Bird': 1,
    'Nick Cassidy': 3,
    'Lucas Di Grassi': 2,
    'Jean-Éric Vergne': 5,
    'Stoffel Vandoorne': 7
}

classificacao = calcular_classificacao(pilotos_posicoes)
exibir_ranking(classificacao)
```

```python
# Quiz game
import pandas as pd

quiz_perguntas = {
    1: {
        'pergunta': "Qual é a velocidade máxima dos carros da Fórmula E?",
        'alternativas': ['a) 220 km/h', 'b) 250 km/h'],
        'resposta_correta': 'b'
    },
    2: {
        'pergunta': "Qual é o ex-piloto de Fórmula 1 que fundou a equipe Venturi Racing na Fórmula E?",
        'alternativas': ['a) Alain Prost', 'b) Gildo Pastor'],
        'resposta_correta': 'b'
    },
    3: {
        'pergunta': "Qual é o nome do circuito de rua em Nova York, onde a Fórmula E realiza uma corrida?",
        'alternativas': ['a) Brooklyn street circuit', 'b) Manhattan street circuit'],
        'resposta_correta': 'a'
    }
}

respostas_usuario = []
corretas = 0

for q_num, q_info in quiz_perguntas.items():
    print(f"{q_num}. {q_info['pergunta']}")
    for alt in q_info['alternativas']:
        print(alt)
    
    resposta = input("Sua resposta (a/b): ").lower()
    respostas_usuario.append(resposta)
    if resposta == q_info['resposta_correta']:
        corretas += 1

total_perguntas = len(quiz_perguntas)
percentual_acertos = (corretas / total_perguntas) * 100
print(f"\nVocê acertou {corretas} de {total_perguntas} perguntas ({percentual_acertos:.2f}% de acertos).")


resultados_quiz = pd.DataFrame({
    'Pergunta': [q_info['pergunta'] for q_info in quiz_perguntas.values()],
    'Resposta Correta': [q_info['resposta_correta'] for q_info in quiz_perguntas.values()],
    'Sua Resposta': respostas_usuario
})

estatisticas = {
    'Total de Perguntas': total_perguntas,
    'Total de Acertos': corretas,
    'Total de Erros': total_perguntas - corretas,
    'Percentual de Acertos': percentual_acertos
}
estatisticas_df = pd.DataFrame([estatisticas])

print("\nEstatísticas do seu desempenho:")
print(estatisticas_df)
```
